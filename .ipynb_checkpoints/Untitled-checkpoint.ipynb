```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x_train is the input variable(size in 1000 squre)

#y_train is the targe(price in 1000s of dollars)

x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])
print(f"x_train is = {x_train}")
print(f"y_train is = {y_train}")
```

```python
# m is the number of training numbers
print(f"x_train.shape:={x_train.shape}")
m = x_train.shape[0]
```

```python
print(f"number of training example is :{m}")
#we also use len function 
m= len(x_train)
print(f"number of training example through lens funtion {m}")
```

```python
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({1}),y^({1})) =  {x_i},{y_i}")
```

```python
#ploting the data 
#ploting data point 
plt.scatter(x_train,y_train,marker='x',c='r')
#set the title 
plt.title("house prize")
#set axis
plt.ylabel('price(in 1000s of dollor)')
plt.xlabel('size(1000s sqrft)')
plt.show()
```

```python
w=100
b= 100
print(f"w:{w}")
print(f"b:{b}")
```

```python
def compute_model_output(x,w,b):
    """compute  the prediction of linear model"""
    m=x.shape[0];
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]= w*x[i]+b
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w,b)
#ploting our model  prediction
plt.plot(x_train,tmp_f_wb,c='b',label='our prediction')
```

```python
#plot the data points
plt.scatter(x_train,y_train,marker = 'x',c='r',label = 'autual values')
#set the title
plt.title("housing prices")
#set axis
plt.ylabel('price(in 1000s of dollors)')
plt.xlabel('size(1000 sqft)')
plt.legend()
plt.show()
```
