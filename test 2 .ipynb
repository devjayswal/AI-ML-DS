```python
#implementing cost function
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0,1.7,2.0,2.5,3.0,3.2])
y_train = np.array([250,300,480,430,630,730])
```

```python
def compute_model_output(x,w,b):
    """compute the prediction of linear model"""
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b
    return f_wb
```

```python
w=195
b=50
tmp_f_wb = compute_model_output(x_train,w,b)
#ploting graph first the in calculate cost function values
plt.plot(x_train,tmp_f_wb,c='b',label='prediction')
#ploting datapoints 
plt.scatter(x_train,y_train,marker='x',c='r',label='actual values')
plt.title("housepricing")
plt.ylabel('price(in 1000 doller)')
plt.xlabel('size(1000 sqft)')
plt.legend()
plt.show()
```

```python
def compute_cost(x,y,w,b):
    """cost function body"""
    #m is the number of training example
    m = x.shape[0]
    cost_sum = 0
    for i in range(m):
        f_wb = w *x[i]+b
        cost = (f_wb -y[i])**2
        cost_sum = cost_sum+cost
        total_cost = (1/(2*m))*cost_sum
    return total_cost
print(f"total_cost_function error value   is = {compute_cost(x_train,y_train,w,b)}")
```
