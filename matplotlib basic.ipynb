```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")

#plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
 x= np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
#make the first plot
plt.plot(x,y_sin,'r--')
plt.title('sine')
#set the second subpoltas active and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cosine')


#show the figure
plt.show()

```

```python
#bar plot
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.ylabel('y axis')
plt.xlabel('x axis ')
plt.show()
```

```python
data = [np.random.normal(0,sld,100) for sld in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True ,patch_artist=True)
```

```python
#pie Chart
label = 'python','c++','Ruby','Java'
sizes =[215,130,245,210]
colors=['gold','yellowgreen','red','lightskyblue']
explode=(0.3,0,0.1,0) #explode 1st slice
#plot
plt.pie(sizes,explode=explode,  colors=colors,autopct='%1.1f%%',shadow= True)
plt.axis('equal')
plt.show()
```

```python
x=np.random.randint(20,high=100,size=(40,1))
y=np.random.randint(30,high=200,size=(40,1))
z=np.random.randint(30,high=250,size=(40,1))
```

```python
from matplotlib import cm
plt.style.use('_mpl-gallery')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)
ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])
plt.show()
```
