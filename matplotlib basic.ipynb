```python
x= np.arange(0,10)
y= np.arange(11,21)
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('matpllotlib demostration')
```

scatter funtion is used for scatter or make points from given vector and c='b' for color of points


```python
#label function is used for 
#make axis label and title for give a title to graph
y= x*x*x + 3*x*x+ 2*x+5
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
```

```python
# we can also crete subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(x,y,'b--')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")

#plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
 x= np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
#make the first plot
plt.plot(x,y_sin,'r--')
plt.title('sine')
#set the second subpoltas active and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cosine')


#show the figure
plt.show()

```

```python
#bar plot
x=[2,8,10]
y=[11,16,9]
x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.ylabel('y axis')
plt.xlabel('x axis ')
plt.show()
```

```python
#histogram
a=np.array([45,75,45,32,88,47,14,13,64,8,85,56,55,74,52,68])
plt.hist(a,bins=25)
plt.title("histogram")
plt.show()
```

## box plot using matplot

```python
data = [np.random.normal(0,sld,100) for sld in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True ,patch_artist=True)
```

```python
#pie Chart
label = 'python','c++','Ruby','Java'
sizes =[215,130,245,210]
colors=['gold','yellowgreen','red','lightskyblue']
explode=(0.3,0,0.1,0) #explode 1st slice
#plot
plt.pie(sizes,explode=explode,  colors=colors,autopct='%1.1f%%',shadow= True)
plt.axis('equal')
plt.show()
```
